```python
import csv

survived_list = []
fare_list = []

with open("titanic.csv", newline="") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    next(reader)

    for line in reader:
        survived_list.append(int(line[0]))
        fare_list.append(float(line[6]))
print(survived_list)
print(fare_list)
[0, 1, 1, 0, 0]


```

```python
import csv
from statistics import mean

# kiek procentu zmoniu isgyveno?
sprob_survival = survived_list.count(1) / len(survived_list)
print(f"Probability of survival in Titanic: {sprob_survival}")

# kokia vidutine kaina Titaniko bilieto?
fare_list = list(filter(lambda x: x > 0 , fare_list)) # filter out all of zero fares
fare_mean = mean(fare_list)
print(f"The mean price of a Titanic tiscket is: {fare_mean:.2f}")
print(f"The min price of a Titanic tiscket is: {min(fare_list):.2f}")
print(f"The max price of a Titanic tiscket is: {max(fare_list):.2f}")

```
